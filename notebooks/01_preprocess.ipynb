```python
import numpy as np
import unicodedata
import string
from tqdm import tqdm
```

```python
# Maximum password length used in the paper
MAX_LEN = 16

# Use 94 printable ASCII characters
VOCAB = list(string.printable[:-6])

# Map characters to integer indices
CHAR_TO_IDX = {c: i+1 for i, c in enumerate(VOCAB)}

# Padding token index
PAD_TOKEN = 0

print("Vocabulary size:", len(VOCAB))
```

```python
def encode_password(p):

    # Convert characters to integer indices
    encoded = [CHAR_TO_IDX[c] for c in p[:MAX_LEN]]

    # Pad shorter passwords
    if len(encoded) < MAX_LEN:
        encoded += [PAD_TOKEN] * (MAX_LEN - len(encoded))

    return encoded
```
