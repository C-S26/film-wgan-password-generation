```python
import numpy as np
```

```python
# Generated samples should be saved after training
samples = np.load("generated_samples.npy")

print("Generated samples:", samples.shape)
```

```python
def uniqueness(samples):

    unique = len(set([tuple(x) for x in samples]))

    return unique / len(samples)

print("Uniqueness:", uniqueness(samples))
```

```python
def get_ngrams(seq, n):

    return [tuple(seq[i:i+n]) for i in range(len(seq)-n+1)]
```

```python
def ngram_coverage(real, generated, n=2):

    real_set = set()
    gen_set = set()

    for r in real:
        real_set.update(get_ngrams(r,n))

    for g in generated:
        gen_set.update(get_ngrams(g,n))

    return len(gen_set & real_set) / len(real_set)
```

```python
real = np.load("train_data.npy")

print("2-gram coverage:", ngram_coverage(real, samples, 2))
print("3-gram coverage:", ngram_coverage(real, samples, 3))
print("4-gram coverage:", ngram_coverage(real, samples, 4))
```
